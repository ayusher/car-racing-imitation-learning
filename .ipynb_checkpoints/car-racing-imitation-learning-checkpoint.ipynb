### Quick and easy imtiation learning!

First cell establishes the RGB to gray conversion function
Second cell lets you play 5 episodes of the car racing game
Third cell trains a Tensorflow model to play like you (then saves to "model/")

```python
import numpy as np

def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
from gym.envs.box2d import *

if __name__ == "__main__":
    from pyglet.window import key

    a = np.array([0.0, 0.0, 0.0])

    def key_press(k, mod):
        global restart
        if k == 0xFF0D:
            restart = True
        if k == key.LEFT:
            a[0] = -1.0
        if k == key.RIGHT:
            a[0] = +1.0
        if k == key.UP:
            a[1] = +1.0
        if k == key.DOWN:
            a[2] = +0.8  # set 1.0 for wheels to block to zero rotation

    def key_release(k, mod):
        if k == key.LEFT and a[0] == -1.0:
            a[0] = 0
        if k == key.RIGHT and a[0] == +1.0:
            a[0] = 0
        if k == key.UP:
            a[1] = 0
        if k == key.DOWN:
            a[2] = 0

    env = CarRacing()
    env.render()
    env.viewer.window.on_key_press = key_press
    env.viewer.window.on_key_release = key_release
    record_video = False
    if record_video:
        from gym.wrappers.monitor import Monitor

        env = Monitor(env, "/tmp/video-test", force=True)
    isopen = True
    states = []
    actions = []
    for i in range(5):
        obs = env.reset()
        old_s = rgb2gray(obs)
        s = obs
        last6 = [old_s for _ in range(6)]
        total_reward = 0.0
        steps = 0
        restart = False
        while True:
            last6.pop(0)
            last6.append(rgb2gray(s))
            states.append(np.dstack(tuple(last6)))
            s, r, done, info = env.step(a)
            total_reward += r
            actions.append(np.array([g for g in a]))
            if steps % 200 == 0 or done:
                print("\naction " + str(["{:+0.2f}".format(x) for x in a]))
                print("step {} total_reward {:+0.2f}".format(steps, total_reward))
            steps += 1
            isopen = env.render()
            if done or restart or isopen == False:
                break
    env.close()
```

```python
import tensorflow as tf
from tensorflow.keras import layers

model = tf.keras.Sequential([   
  layers.Conv2D(32, (5, 5), activation='relu', input_shape=(96, 96, 6)),
  layers.MaxPooling2D(),

  layers.Conv2D(32, (5, 5), activation='relu'),
  layers.MaxPooling2D(),

  layers.Conv2D(32, (3, 3), activation='relu'),
  layers.MaxPooling2D(),
    
  layers.Conv2D(32, (3, 3), activation='relu'),
  layers.MaxPooling2D(),

  layers.Flatten(),
  layers.Dense(512, activation = 'tanh'),
  layers.Dropout(.2),
  layers.Dense(3)
])

model.compile(
  optimizer='adam',
  loss=tf.losses.mean_squared_error,
  metrics=['accuracy'])

model.summary()

model.fit(
  np.array(states)/255.0,
  np.array(actions),
  epochs=5
)

model.save("model")
```
